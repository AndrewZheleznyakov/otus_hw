# I. Numpy

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
arr = np.zeros(10)
arr[5] = 1
arr = arr.reshape((2,5))
print (arr)
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
arr = np.arange(10,50)[::-1]
print (arr[1::2])
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
arr = np.arange(9).reshape(3,3)
print (arr)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
arr = np.random.rand(4,3,2)
print(arr)
print("max: ",arr.min())
print("min: ",arr.max())
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
arr = np.random.randint(10,size=(6,4))
arr2 = np.random.randint(10,size=(4,3))

print("arr:\n",arr)
print("arr2:\n",arr2)

print("\n\n\n", np.dot(arr,arr2))
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
arr = np.random.randint(10,size=(7,7))

print(arr)

print(np.std(arr))
print("var:",np.var(arr))

arrmax, arrmin = arr.max(), arr.min()

arr = (arr - arrmin)/(arrmax - arrmin)

print("normalized: \n",arr)
```

```python
len(ds.loc[ds['smoker'] == 'Yes'])
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
days = ds['day'].unique()
test = list(map(lambda x: ds.loc[ds['day'] == x]['total_bill'].mean(), days))
print(list(days))
print (test)
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
newDs = ds[ds['total_bill'] > ds['total_bill'].median()]

sex = ds['sex'].unique()
test = list(map(lambda x: newDs.loc[newDs['sex'] == x]['tip'].mean(), sex))

print(list(sex))
print (test)
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
ds['smoker'] = [0 if i == 'No' else 1 for i in ds['smoker']]
```
